## Data wrangling

```python
df1=df1.dropna(subset=['mixins'])
df1
```

I already dropped irrelevant columns and remove nulls

```python
df1.reset_index(drop=True, inplace=True)
```

```python
#create columns for different mixins
df1['chocolate'] = 0
df1['raisins'] = 0
df1['oats'] = 0
df1['nuts'] = 0
df1['peanut butter'] = 0
```

```python
df1.head()
```

```python
#fill these columns and afterwards drop column 'mixins'
mixins = ['chocolate', 'raisins', 'oats', 'nuts', 'peanut butter']
for mix in mixins:
    for i in range(len(df1)):
        if mix in [x.strip() for x in df1.loc[i, 'mixins'].split(',')]:
            df1.loc[i, mix] = 1
```

```python
df1.drop(columns=['mixins'], inplace=True)
```

```python
df1.head(100)
```

```python
#plot new correlation heatmap
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(df1.corr(), dtype=np.bool))
heatmap = sns.heatmap(df1.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='plasma_r')
```

### Pre-processing

```python
df1.dtypes
```

```python
#only categorical: 'butter type'
df1 = pd.get_dummies(df1, columns=['butter type'])
```

```python
df1
```

```python
df1.info()
```

```python
df1.isna().sum()
```

```python
df1.reset_index(drop=True, inplace=True)
```

### Split data into dependent and independent variables

```python
#want to predict the quality
X= df1.drop(columns='quality')
y= df1.quality
```

### Pre-processing : scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X_scaled, y, test_size=0.1, random_state=23)
```

### Model, prediction and accuracy

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import math
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
```

```python
y_pred  = lm.predict(X_test)
```

```python
#accuracy metrics
print('R2 score = ', r2_score(y_test, y_pred))
print('MSE = ', mean_squared_error(y_test, y_pred))
print('RMSE = ', math.sqrt(mean_squared_error(y_test, y_pred)))
```

# Can we get more accuracy with an ensemble method?

#### Add a random forest regressor model to try to get a more accurate score.

A Random Forest is an ensemble technique capable of performing both regression and classification tasks with the use of multiple decision trees and a technique called Bootstrap and Aggregation, commonly known as bagging. The basic idea behind this is to combine multiple decision trees in determining the final output rather than relying on individual decision trees.

```python
#want to predict the quality
X= df1.drop(columns='quality')
y= df1.quality
```

```python
X.shape
```

```python
y.shape
```

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
  
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 23)
  
# fit the regressor with x and y data
regressor.fit(X_train, y_train)  
```

```python
y_predicted = regressor.predict(X_test)
y_predicted
```

Evaluate accuracy:

```python
#accuracy metrics
print('R2 score = ', r2_score(y_test, y_predicted))
print('MSE = ', mean_squared_error(y_test, y_predicted, squared=False))
print('RMSE = ', math.sqrt(mean_squared_error(y_test, y_predicted, squared=False)))
```

The RMSE is lower, so the accuracy has improved.
